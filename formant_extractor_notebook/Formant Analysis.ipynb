```python
from load_chopped_clip import *
from plt_signal import *
from speech_analysis_helpers import compute_stft, find_peaks, lpc_analysis
from local_constants import CURR_DIR_PATH
```

```python
# Declare constants
sample_rate = 10000
```

```python
# Get a sample speech file, of the 'ah' vowel

orig_clip = load_clip("{}/sound_clips/{}".format(CURR_DIR_PATH, "m01ae.wav"), sample_rate)
chopped_clip = load_chopped_clip("{}/sound_clips".format(CURR_DIR_PATH), "m01ae.wav", sample_rate)

# Visualize the original and chopped clips
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
plt_signal(orig_clip, sample_rate, player=True)
plt.subplot(1,2,2)
plt_signal(chopped_clip, sample_rate, player=True)
```

```python
# Compute the STFT of the signal
# NOTE: vp stands for 'voiceprint'
f, t, vp_stft = compute_stft(chopped_clip, sample_rate, visualize=True)
```

```python
# Interpolate the 10th 'slice' of the LPC
stft_slice_index = 10
main_vp_stft_slice_interp = find_peaks(f, t, vp_stft, stft_slice_index, visualize=True, figsize=(30, 5), lin_interp=True)
```

```python
# Compute the formants for the LPC. 
# Below, we see a reconstruction of the chopped clip via LPC method,
# and we print the associated formants of the signal.
formants = lpc_analysis(chopped_clip, sample_rate, visualize=True)

print(formants)
```
